```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("All_IPL_Match_Scores_raw_backup.xlsx")
df
```

```python
delhi = df[df["Winner"] == "Delhi"].groupby("Year").Year.count()
delhi
rajasthan = df[df["Winner"] == "Rajasthan"].groupby("Year").Year.count()
rajasthan
chennai = df[df["Winner"] == "Chennai"].groupby("Year").Year.count()
chennai
```

```python
plt.plot(delhi, marker="+", markersize=10, linewidth=2, linestyle="-.", label='Delhi')
plt.plot(rajasthan, marker="*", markersize=10, linewidth=2, linestyle="-", label='Rajasthan')
plt.plot(chennai, marker="o", markersize=10, linewidth=2, linestyle=":", label='Chennai')

plt.legend()

plt.xlabel("Year")
plt.ylabel("How many matches won by team")
plt.title("Matches won by Delhi, Rajasthan & Chennai")
```

```python
plt.hist(delhi, label="Delhi")
plt.hist(rajasthan, label="Rajasthan")
plt.hist(chennai, label="Chennai")

plt.legend()
```
